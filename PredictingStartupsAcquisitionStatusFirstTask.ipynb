# Filling the missing values in the latitude and longitude columns


```python
df_new['lng'] = df_new['lng'].interpolate(method='linear')
df_new['lat'] = df_new['lat'].interpolate(method ='linear')
```

#Filling the missing values in Numerical columns

```python
num_cols = ['logo_width','logo_height','funding_rounds','funding_total_usd','milestones','relationships']

for i in num_cols:
    df_new[i].fillna(df_new[i].mean(),inplace = True)
```

Extracting Day, Month and Year from Date columns

```python
for col in df_new.columns:
    if df_new[col].dtype == 'datetime64[ns]':
        df_new[f'{col}_Day'] = df_new[col].dt.day
        df_new[f'{col}_Month'] = df_new[col].dt.month
        df_new[f'{col}_Year'] = df_new[col].dt.year
```

```python
#some outliers are present in logo_width ,logo_height,relationships that should be replaces with median

columns_with_outliers = ['logo_width','logo_height','relationships']

for i in columns_with_outliers:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df[i] < (Q1 - 1.5 * IQR)) | (df[i] > (Q3 + 1.5 * IQR)))
    median = df[i].median()
    df.loc[outliers, i] = median
```
